```python
def calc_coeffs(n_start,n_end,size):
    size = int(size)
    x = np.linspace(-np.pi,np.pi,size)
    sx = np.abs(x)
    dx = (x[-1]-x[0])/size
    an,bn = [],[]
    for n in range(n_start,n_end):
        x = np.linspace(-np.pi,np.pi,size)
        sx = np.abs(x)
        cosnx = np.cos(n*x)
        sinnx = np.sin(n*x)
        an.append(1/np.pi*np.sum(sx*cosnx*dx))
        bn.append(1/np.pi*np.sum(sx*sinnx*dx))
    return an,bn

def calc_series(an,bn,n_start,n_end,size):
    size = int(size)
    x = np.linspace(-np.pi,np.pi,size)
    cosnx = np.array([np.cos(n*x) for n in range(n_start,n_end)])
    sinnx = np.array([np.sin(n*x) for n in range(n_start,n_end)])
    an = np.array(an)
    bn = np.array(bn)
    return np.sum(np.dot(an,cosnx))+np.sum(np.dot(bn,sinnx))
```

```python
an,bn = calc_coeffs(-10,10,1e4)
calc_series(an,bn,-10,10,1000)
```
