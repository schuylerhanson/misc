1. Modeling relationship between $\Delta SR$ and $\Delta P/E$ from "Common Sense Investing" using linear regression

```python
# Reading in Bogle's data (p.16, exhibit 2.2)
X = np.array([-4,3.7,.4,-6.1,10.3,-1.6,-8.5,7.9,14.5,-7.9])
Y = np.array([-3.4,3.3,.3,-6.3,9.3,-1,-7.5,7.7,7.2,-3])
```

```python
# Estimators
# Use (un)biased sample variance estimator?
mu_X = np.mean(X)
mu_Y = np.mean(Y)
ssX = (np.sum((1/9)*(X-mu_X)**2))**.5
ssY = (np.sum((1/9)*(Y-mu_Y)**2))**.5
sX = np.var(X)**.5
sY = np.var(Y)**.5

# covariance, correlation
# As corr ~ 1, Y = aX+b
cov = sum([.1*x*y for x,y in zip(X-mu_X,Y-mu_Y)])
corr_biased = cov/(sX*sY)
corr_unbiased = cov/(ssX*ssY)
corr_biased,corr_unbiased
```

```python
# Constructing linear model Y_est = B1*X+B0
# c.f. discussion in Wasserman for equations
num = sum([x*y for x,y in zip(X-mu_X,Y-mu_Y)])
den = sum([x**2 for x in X-mu_X])
B1 = num/den
B0 = mu_Y-B1*mu_X
```

```python
plt.plot(X,Y,'.')
plt.plot(X,B0+B1*X)
```

```python
# Expected relative error: E[e_est,rel]
Y_est = B0+B1*X
e_rel = (Y_est-Y)/Y
mu_erel = np.mean(e_rel)
```
