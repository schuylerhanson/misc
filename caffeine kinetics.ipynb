```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A0 = [230,180,130,90]
k = np.log(2)/4.5
t = np.linspace(0,13,1000)
A1 = []
for A in A0:
    A1.append(A*np.exp(-k*t))
fig,ax = plt.subplots()
for A,x,a0 in zip(A1,[50,100,150,200],A0):
    ax.plot(t,A,label=a0)
    ax.axhline(y=x,color='k',linestyle='dotted')
    ax.legend()
ax.set_xlabel('t [h]')
ax.set_ylabel('mg')
```
